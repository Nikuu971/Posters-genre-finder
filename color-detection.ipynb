<h3 style="color:black">Affichage prediction</h3>

```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import csv
import cv2
import urllib.request as urllib
import numpy as np

file = open("dataset/imdb_top_1000.csv",encoding="utf8")
csvreader = csv.reader(file)
rows = []
rows2=[]
samples_x=[]
samples_y=[]
model=SGDClassifier()

for row in csvreader:
    rows.append(row[0])
    rows2.append(row[5])
    
i=1
while i<len(rows):
    try :
        urllib.urlopen(rows[i])
    except :
        i=i+1
    
    res=urllib.urlopen(rows[i])
    image=np.asarray(bytearray(res.read()),dtype="uint8")
    image=cv2.imdecode(image, cv2.IMREAD_COLOR)
    
    height,width, _ =np.shape(image)
    
    data = np.reshape(image, (height * width, 3))
    data = np.float32(data)
    
    number_clusters = 5
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
    flags = cv2.KMEANS_RANDOM_CENTERS
    compactness, labels, centers = cv2.kmeans(data, number_clusters, None, criteria, 10, flags)
    #print(centers)
    genre=rows2[i]
    samples_x.append(centers)
    samples_y.append(genre)
    i=i+1

x=np.array(samples_x)
y=np.array(samples_y)

x = x.reshape(x.shape[0], (x.shape[1]*x.shape[2]))

x_train, x_test, y_train, y_test = train_test_split(x, y)

preds=model.fit(x_train,y_train).predict(x_test)

#model.score(x_test,y_test)

accuracy_score(y_test, preds)
 
print(classification_report(y_test, preds))

file.close()
```
